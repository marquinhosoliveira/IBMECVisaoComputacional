```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import os
from skimage import io
from google.colab.patches import cv2_imshow
from skimage.filters import threshold_otsu
from skimage.filters import try_all_threshold
```

Exercício 01: Ler uma imagem hospedada na Internet e exibi-la no projeto. (0,30 ponto)

```python
origem = origem = "https://storage.googleapis.com/dpw/app/uploads/2009/12/como-otimizar-imagens-web-internet1.jpg"
image = io.imread(origem) 
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
cv2_imshow(img)
```

```python
print(f"(Altura, Largura, Qtde de Canais): {img.shape}") 
```

Exercício 03: Redimensionar a imagem para uma resolução de 200 X 200 pixels. (0,30 ponto)

```python
def redimensionar_imagemTamanho(altura, largura, img):
  dim = (largura, altura)  
  img_redimensionada = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
  return img_redimensionada 
```

```python
img_200x200 = redimensionar_imagemTamanho(200,200, img)
cv2_imshow(img_200x200)
print(f"(Altura, Largura, Qtde de Canais): {img_200x200.shape}") 
```

Exercício 04: Aplicar um filtro Gaussiano na imagem do exercício 03 e exibir o resultado. (0,30 ponto)

```python
kernel = np.array([[1, 2, 1], 
                   [1, 1, 3], 
                   [1, 4, 1], 
                   ])
kernel = kernel/sum(kernel)
img_passa_baixa = cv2.filter2D(img_200x200,
                               -1,
                               kernel)
cv2_imshow(img_passa_baixa)
```

Exercício 05: Aplicar uma limiarização na imagem do exercício 03 e exibir o resultado. (0,30 ponto)

```python
#Converter para cinza
img_cinza=cv2.cvtColor(img_200x200,cv2.COLOR_RGB2GRAY)
cv2_imshow(img_cinza)
```

```python
limiar = threshold_otsu(img_cinza)
print(limiar)
img_otsu  = img_cinza < limiar
img_filtrada = img_cinza*img_otsu
cv2_imshow(img_filtrada)
```
